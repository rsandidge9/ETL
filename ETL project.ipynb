```python
import pandas as pd
```

```python
f15=pd.read_csv("datasets_894_813759_2015.csv")
f16=pd.read_csv("datasets_894_813759_2016.csv")
f17=pd.read_csv("datasets_894_813759_2017.csv")
f18=pd.read_csv("datasets_894_813759_2018.csv")
f19=pd.read_csv("datasets_894_813759_2019.csv")
f17.head()
```

```python
f15=f17[["Country","Happiness.Rank","Happiness.Score"]]
f16=f17[["Country","Happiness.Rank","Happiness.Score"]]
f17=f17[["Country","Happiness.Rank","Happiness.Score"]]
f18=f17[["Country","Happiness.Rank","Happiness.Score"]]
f19=f17[["Country","Happiness.Rank","Happiness.Score"]]
f15.set_index("Country")
f16.set_index("Country")
f17.set_index("Country")
f18.set_index("Country")
f19.set_index("Country")

f15.head()
```

```python
free_econ=pd.read_csv("economic freedom data.csv")
free_econ=free_econ[["Name","Index Year","Overall Score"]]
free_econ=free_econ.rename(columns={'Name':'Country'})
free_econ.set_index('Country')
free_econ.head()
```

```python
free_econ17=free_econ.loc[free_econ["Index Year"]==2017]
free_econ16=free_econ.loc[free_econ["Index Year"]==2016]
free_econ15=free_econ.loc[free_econ["Index Year"]==2015]
free_econ18=free_econ.loc[free_econ["Index Year"]==2018]
free_econ19=free_econ.loc[free_econ["Index Year"]==2019]

free_econ19.head()
```

```python
df15=pd.merge(f15, free_econ15, how="inner", on="Country")
df15=df15.rename(columns={"Overall Score":"Econ Freedom"})

df16=pd.merge(f16, free_econ16, how="inner", on="Country")
df16=df16.rename(columns={"Overall Score":"Econ Freedom"})

df17=pd.merge(f17, free_econ17, how="inner", on="Country")
df17=df17.rename(columns={"Overall Score":"Econ Freedom"})

df18=pd.merge(f18, free_econ18, how="inner", on="Country")
df18=df18.rename(columns={"Overall Score":"Econ Freedom"})

df19=pd.merge(f19, free_econ19, how="inner", on="Country")
df19=df19.rename(columns={"Overall Score":"Econ Freedom"})

df19.head()
```
