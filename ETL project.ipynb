```python
import pandas as pd
import bs4 as bs
from bs4 import BeautifulSoup
import requests
import urllib
import urllib.request
from splinter import Browser
```

```python
f15=pd.read_csv("datasets_894_813759_2015.csv")
f16=pd.read_csv("datasets_894_813759_2016.csv")
f17=pd.read_csv("datasets_894_813759_2017.csv")
f18=pd.read_csv("datasets_894_813759_2018.csv")
f19=pd.read_csv("datasets_894_813759_2019.csv")
#f20=pd.read_csv("2020_happy_report_2.csv")
#f15.head()
f19.head()
```

```python
f15=f15[["Country","Happiness Rank","Happiness Score"]]
f16=f16[["Country","Happiness Rank","Happiness Score"]]
f17=f17[["Country","Happiness.Rank","Happiness.Score"]]
f18=f18[["Country or region","Overall rank","Score"]]
f19=f19[["Country or region","Overall rank","Score"]]
f17=f17.rename(columns={'Happiness.Rank':'HappinessRank',"Happiness.Score":"Happiness Score"})
f18=f18.rename(columns={"Country or region":"Country",'Overall rank':'Happiness Rank',"Score":"Happiness Score"})
f19=f19.rename(columns={"Country or region":"Country",'Overall rank':'Happiness Rank',"Score":"Happiness Score"})

f15.set_index("Country")
f16.set_index("Country")
f17.set_index("Country")
f18.set_index("Country")
f19.set_index("Country")

f19.head()
```

```python
free_econ=pd.read_csv("economic freedom data.csv")
free_econ=free_econ[["Name","Index Year","Overall Score"]]
free_econ=free_econ.rename(columns={'Name':'Country'})
free_econ.set_index('Country')
free_econ.head()
```

```python
free_econ17=free_econ.loc[free_econ["Index Year"]==2017]
free_econ16=free_econ.loc[free_econ["Index Year"]==2016]
free_econ15=free_econ.loc[free_econ["Index Year"]==2015]
free_econ18=free_econ.loc[free_econ["Index Year"]==2018]
free_econ19=free_econ.loc[free_econ["Index Year"]==2019]

free_econ19.head()
```

```python
df15=pd.merge(f15, free_econ15, how="inner", on="Country")
df15=df15.rename(columns={"Overall Score":"Econ Freedom"})

df16=pd.merge(f16, free_econ16, how="inner", on="Country")
df16=df16.rename(columns={"Overall Score":"Econ Freedom"})

df17=pd.merge(f17, free_econ17, how="inner", on="Country")
df17=df17.rename(columns={"Overall Score":"Econ Freedom"})

df18=pd.merge(f18, free_econ18, how="inner", on="Country")
df18=df18.rename(columns={"Overall Score":"Econ Freedom"})

df19=pd.merge(f19, free_econ19, how="inner", on="Country")
df19=df19.rename(columns={"Overall Score":"Econ Freedom"})

df19.head()
```

```python
import requests
from api_keys import weather_api_key
api_key=weather_api_key
#print(api_key)
from citipy import citipy
city_name=[]
city_weather=[]
#because we need to pay for the historical data, so just use  current weather for this project
#url="https://history.openweathermap.org/data/2.5/aggregated/year?"+ api_key
#response = requests.get(f"{url}&q={city}&appid=").json()
for city in cty:

    try:
        url="http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key
        response = requests.get(f"{url}&q={city}").json() 
        name=response["name"]
        weather=response['main']['feels_like']
        city_name.append(name)
        city_weather.append(weather)
        #print(response)
    except:
        # Append null values
        print("City not found")
        pass

print(city_name) 
print("The feels like temp for the above city")
print(city_weather)
```
